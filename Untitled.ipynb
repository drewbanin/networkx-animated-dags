```python
%matplotlib inline

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML, Image

rc('animation', html='html5')
```

```python
from networkx.drawing.nx_agraph import graphviz_layout
from networkx.drawing.nx_agraph import to_agraph


graph = nx.DiGraph()

graph.add_node("a", cost=3)
graph.add_node("b", cost=5)
graph.add_node("c", cost=2)
graph.add_node("d", cost=2)
graph.add_node("e", cost=5)
graph.add_node("f", cost=2)
graph.add_node("g", cost=2)
graph.add_node("h", cost=3)
graph.add_node("i", cost=2)
graph.add_node("j", cost=2)

graph.add_edge("a", "b")
graph.add_edge("a", "c")
graph.add_edge("a", "d")

graph.add_edge("b", "e")
graph.add_edge("b", "f")
graph.add_edge("c", "g")
graph.add_edge("d", "g")

graph.add_edge("e", "h")
graph.add_edge("g", "i")
graph.add_edge("i", "j")

for node in graph.node:
    ancestors = nx.ancestors(graph, node)
    cost = graph.node[node]['cost']
    in_deg = graph.in_degree[node]
    for ancestor in ancestors:
        cost += graph.node[ancestor]['cost']
    graph.node[node]['true_cost'] = cost
    graph.node[node]['depth'] = len(nx.shortest_path(graph, 'a', node))

pos = graphviz_layout(graph, prog='dot', args='-Grankdir="LR" -Granksep="10" ')
labels = {n: "{} ({})".format(n, graph.node[n]['cost']) for n in graph.node}

color_map = {n: '#cccccc' for n in graph.node}
def colors():
    return [color_map[n] for n in graph.node]

max_degrees = 0

# Build plot
fig, ax = plt.subplots(figsize=(6,4))

max_degrees = 0
node_progress = {}
nodes_to_process = set()

def update(i):
    global nodes_to_process, max_degrees
    
    if len(nodes_to_process) == 0:
        nodes_to_process = {n for n in graph.node if graph.node[n]['depth'] == max_degrees}
        max_degrees += 1
        
        for node in nodes_to_process:
            color_map[node] = '#ff0000'
            node_progress[node] = 0
    
    completed_nodes = []
    if i > 0:
        for node in nodes_to_process:
            node_progress[node] += 1
            if node_progress[node] >= graph.node[node]['cost']:
                color_map[node] = '#00ff00'
                completed_nodes.append(node)

    for n in completed_nodes:
        nodes_to_process.remove(n)
            

    ax.clear()
    nx.draw(graph,
            pos=pos,
            node_size=1500,
            node_color=colors(),
            arrows=True,
            node_shape="o",
            labels=labels,
            with_labels=True,
            ax=ax)
    ax.set_title("Time={}".format(i))
    ax.set_xticks([])
    ax.set_yticks([])
            

animation = matplotlib.animation.FuncAnimation(fig, update, frames=20, interval=500, repeat=False)
```
